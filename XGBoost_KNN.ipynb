```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.preprocessing import StandardScaler,LabelEncoder 
from sklearn.svm import SVC 
from sklearn.metrics import accuracy_score,classification_report , confusion_matrix 
import joblib
import xgboost as xgb
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
```

```python
df = pd.read_csv(url, sep=';')
```

```python
df.head()
```

```python
df["quality"] = df["quality"].apply(lambda x : 1 if x >= 7 else 0)
```

```python
x = df[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]]
```

```python
y = df["quality"]
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train,y_test = train_test_split(x_scaled, y, test_size= .2, random_state=35)
```

```python
xgb_classifier = xgb.XGBClassifier(objective="binary:logistic", eval_metric = "logloss")
```

```python
xgb_classifier.fit(x_train,y_train)
```

```python
xgb_classifier.predict(x_test)
```

```python
accuracy_score(y_test, xgb_classifier.predict(x_test) )
```

```python
print("classification_report \n", classification_report(y_test, xgb_classifier.predict(x_test)))
```

```python
confusion_matrix(y_test, xgb_classifier.predict(x_test))
```

```python
#hyperparameter tuning
```

```python
param_grid = {
    "n_estimators":[50,100,200],
    "learning_rate":[0.01,.1,.2],
    "max_depth":[3,5,7,8,9,11],
    "subsample":[.7,.8,1],
    "colsample_bytree":[0.7,0.8,1]
}
```

```python
grid_search = GridSearchCV(xgb.XGBClassifier(objective="binary:logistic", eval_metric = "logloss"),param_grid,cv=3, n_jobs=1, verbose=2)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
xgb_classifier_best_param = xgb.XGBClassifier(**grid_search.best_params_,objective="binary:logistic", eval_metric = "logloss")
```

```python
xgb_classifier_best_param.fit (x_train,y_train)
```

```python
xgb_classifier_best_param.predict(x_test)
```

```python
accuracy_score(y_test, xgb_classifier_best_param.predict(x_test))
```

```python
print(classification_report(y_test, xgb_classifier_best_param.predict(x_test)))
```

```python
confusion_matrix(y_test, xgb_classifier_best_param.predict(x_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
```

```python
irish = load_iris()
```

```python
x = irish.data
y = irish.target
```

```python
df = pd.DataFrame(x,columns=irish.feature_names)
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_train_knn, x_test_knn, y_train_knn, y_test_knn = train_test_split(x_scaled,y,test_size=.2,random_state=35)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train_knn,y_train_knn)
```

```python
knn.predict(x_test_knn)
```

```python
accuracy_score(y_test_knn,knn.predict(x_test_knn))
```

```python
param = {"n_neighbours":np.range(1,100)}
grid_knn = GridSearchCV(KNeighborsClassifier(), param,cv=5,scoring="accuracy",n_jobs=1, verbose=1)
grid_knn.fit(x_train_knn,y_train_knn)
```
